```python
import json
from IPython.display import Image, display

payload = json.loads((REPO_ROOT / 'logs' / 'training_last.json').read_text())
print('Final random avg return:', round(payload['final_eval']['random']['avg_return'], 3))
print('Final trained mean return:', round(payload['final_eval']['llm']['mean_return'], 3))
print('Parse OK rate:', round(payload['final_eval']['llm']['parse_ok_rate'], 3))

display(Image(filename=str(REPO_ROOT / 'docs' / 'figures' / 'training_vs_baselines.png')))
display(Image(filename=str(REPO_ROOT / 'docs' / 'figures' / 'final_random_vs_trained.png')))
```
